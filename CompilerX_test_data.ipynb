### Extracted the data from list value in Industry column

```python
import pandas as pd

def extract_industries(df):
    # Created an empty list to store extracted data
    extracted_data = []
    
    # Iterated over each row in the DataFrame
    for index, row in df.iterrows():
        company = row['company']
        industries = row['industry']
        
        # If industries is a list, extracted each industry
        if isinstance(industries, list):
            for industry in industries:
                extracted_data.append({'company': company, 'industry': industry})
    
    # Converted the list of dictionaries into a DataFrame
    extracted_df = pd.DataFrame(extracted_data)
    
    return extracted_df
```

```python
extracted_df = extract_industries(industry_df)
extracted_df.head()
```

### Extracted the data from dict value in additional_information column

```python
import pandas as pd

def extract_additional_info(df):
    extracted_info = []
    
    for _, row in df.iterrows():
        company_info = {}
        company_info['company'] = row['company']
        
        additional_info = row['additional_information']
        if isinstance(additional_info, dict):
            for key, value in additional_info.items():
                company_info[key] = value
        
        extracted_info.append(company_info)
    
    return pd.DataFrame(extracted_info)
```

```python
extracted_df = extract_additional_info(add_info)
extracted_df.head(10)
```

### Created function to replace and categorize values

```python
def replace_country(value):
    if pd.isnull(value) or value == '' or isinstance(value, float):
        return 'No Data Found'
    elif 'US' in value or 'America' in value or 'Coast' in value or 'Lakes' in value or 'Scandinavia' in value or 'Las Vegas' in value:
        return 'United States'
    elif 'APAC' in value:
        return 'APAC'
    elif 'EU' in value:
        return 'European Union'
    elif 'EMEA' in value:
        return 'EMEA'
    elif 'Gulf Cooperation Council (GCC)' in value or 'India' in value:
        return 'Asia'
    else:
        return 'No Data Found'

def replace_cat_values(value):
    if pd.isnull(value) or value == '' or isinstance(value, float):
        return 'No Data Found'
    else:
        return value

def replace_cat_type(value):
    if pd.isnull(value) or value == '' or isinstance(value, float):
        return 'No Data Found'
    elif value in ('Privately Held', 'Private'):
        return 'Private'
    elif value in ('Non-profit', 'Nonprofit'):
        return 'Non-profit'
    elif value in ('Public Company', 'Public'):
        return 'Public'
    else:
        return value

def replace_no_of_emp(value): 
    if value == '' or pd.isnull(value) or isinstance(value, float): 
        return 'No Data Found'
    elif value in ('1-10', '11-50', '501-1000', '1001-5000', '1001-5000', '5001-10000'): 
        return value
    elif value in ('51-100', '101-250', '251-500'): 
        return '51-500'
    else:
        return '10001+'
```

```python
cat_columns_to_process = [
    'operating_status', 
    'hub_tags', 
    'investment_stage',  
    'funding_status', 
    'top_5_investors', 
    'ipo_status', 
    'stock_exchange'
]
for column in cat_columns_to_process:
    extracted_df[column] = extracted_df[column].apply(replace_cat_values)

extracted_df['headquarters_regions'] = extracted_df['headquarters_regions'].apply(replace_country)
extracted_df['type'] = extracted_df['type'].apply(replace_cat_type)
extracted_df['number_of_employees'] = extracted_df['number_of_employees'].apply(replace_no_of_emp)

extracted_df.head()
```
